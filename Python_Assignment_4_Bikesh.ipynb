**7. Simple Calculator (Conditions Only)**

Take Two number and an operator (+,-,*,/) as input.
Use if-elif-ellse to perform the correct operation

```python
num1 = int (input("Enter a first number : "))
num2 = int(input("Enter a second number : "))

operator = input("Enter a operator (+,-,*,/) :")

#perform simple calculator

if operator == "+":
    add = num1 + num2
    print("The operation is addition")
    print(f"The result is : {add}")
elif operator == "-":
    sub = num1 - num2 
    print("The operation is Subtraction")
    print(f"The result is : {sub}")
elif operator == '*':
    multi = num1 *num2
    print("The operation is multipication")
    print(f"The result is : {multi}")
elif operator == "/":
    div = num1/num2 
    print("The operation is Division")
    print(f"The result is : {div}")
else:
    print("Please choose correct operator!")
```
