### 5 Theory Question - Variables and Data Types
##### 1. What is a variable in Python?

A variable is a name used to store data in a program. It allows us to save values so we can use them later.

##### 2. What are the main data types in Python?

Some main data types in Python are:
Integer (int)
Float (float)
String (str)
Boolean (bool)

##### 3. What is the difference between an integer and a float?

An integer is a whole number (no decimal), like 5 or -10.
A float is a number with a decimal point, like 3.5 or 2.0.

##### 4. What is a string? Give two examples.

A string is a sequence of characters used to store text.
Examples:
"Hello"
"Python Programming"

##### 5. What is a boolean data type used for?

A boolean data type is used to store True or False values.
It is commonly used for making decisions in programs (like checking conditions).
